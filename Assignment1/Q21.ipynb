```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import statistics
```

```python
df = pd.read_csv('Cars.csv')
df
```

```python
nd = stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3],df.iloc[:,4])
```

```python
stat, p = stats.f_oneway(nd,df['MPG'])
if p < 0.05:
    print('Reject Null Hypothesis. Not Normal')
else:
    print('Fail to reject Null hypothesis. Distribution is normal')
```

```python
df = pd.read_csv('wc-at.csv')
df
```

```python
nd = stats.f_oneway(df.iloc[:,0],df.iloc[:,1])
```

```python
stat, p = stats.f_oneway(nd,df['Waist'],df['AT'])
if p < 0.05:
    print('Reject Null Hypothesis. Not Normal')
else:
    print('Fail to reject Null hypothesis. Distribution is normal')
```
