```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy.cluster import hierarchy
from sklearn.cluster import KMeans , DBSCAN
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import AgglomerativeClustering
import numpy as np
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('crime_data.csv',index_col=0)
```

```python
df
```

```python
sns.pairplot(df)
```

# Standardization

```python
sc = StandardScaler()
scaled = sc.fit_transform(df)
```

```python
scaled
```

# Hierarchical Clustering

```python
lk = hierarchy.linkage(scaled,method='single')
dendo = hierarchy.dendrogram(lk)
plt.axhline(1.50,linestyle='--')
plt.show();
```

```python
lk_c = hierarchy.linkage(scaled,method='complete')
dendo = hierarchy.dendrogram(lk_c)
plt.axhline(4,linestyle='--')
plt.show();
```

```python
lk_a = hierarchy.linkage(scaled,method='average')
dendo = hierarchy.dendrogram(lk_a)
plt.axhline(2,linestyle='--')
plt.show();
```

```python
lk_ct = hierarchy.linkage(scaled,method='centroid')
dendo = hierarchy.dendrogram(lk_ct)
plt.axhline(1.750,linestyle='--')
plt.show();
```

# Model

```python
hc = AgglomerativeClustering(n_clusters=4)#since complete methods shows best cluster and it givesus 4 partitions
y = hc.fit_predict(scaled)
```

```python
y
```

```python
df['cluster_h']=y
df
```

```python
df.groupby('cluster_h').agg('mean')
```

# visualizatiom

```python
sns.scatterplot(x=df['Murder'],y=df['Assault'], hue=df['cluster_h'],palette='rainbow')
plt.show()
```

```python
sns.scatterplot(x=df['Rape'],y=df['UrbanPop'], hue=df['cluster_h'],palette='rainbow')
plt.show()
```

# Kmean

```python
km = KMeans()
y = km.fit_predict(scaled)
```

```python
km.cluster_centers_
```

```python
plt.scatter(x=scaled[:,0],y=scaled[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
silhouette_score(scaled,y)
```

# Elbow Graph

```python
wcss = []
for i in range (1,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(scaled)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,20),wcss)
plt.title('Elbow Graph')
plt.axvline(4, linestyle='--',c='r')
plt.show()
```

# Kmeans Final

```python
fkmean = KMeans(n_clusters=4)
ypred = fkmean.fit_predict(scaled)
```

```python
silhouette_score(scaled,ypred)
```

```python
plt.scatter(x=scaled[:,0],y=scaled[:,1],c=ypred,cmap='rainbow')
plt.scatter(fkmean.cluster_centers_[:,0],fkmean.cluster_centers_[:,1],color = 'black',s=100)
plt.show()
```

# Evaluation

# Hyperparameter Tuning

```python
neigh = NearestNeighbors(n_neighbors=3).fit(scaled)
d,i= neigh.kneighbors(scaled)
```

```python
pd.DataFrame(d)
```

```python
dist = np.sort(d[:,1])
```

```python
plt.plot(dist)
plt.xlabel('Distance')
plt.ylabel('Epsilon Value')
plt.axhline(1.3,linestyle='--',c='r')

plt.show()
```
