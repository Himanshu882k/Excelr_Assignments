```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
from wordcloud import WordCloud
import spacy
from nltk.sentiment import SentimentIntensityAnalyzer
import seaborn as sns
sns.set_theme(style='darkgrid',palette='rainbow')
```

```python
df = pd.read_csv('Elon_musk.csv',encoding='latin-1',index_col=0)
```

```python
df
```

# Cleaning

```python
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)

    # Remove mentions (@username)
    text = re.sub(r"@[^\s]+", "", text)

    # Remove hashtags (#hashtag)
    text = re.sub(r"#([^\s]+)", r"\1", text)
    
    # Remove characters inside brackets (<text>, {text}, [text], (text))
    text = re.sub(r"<[^>]+>|\{[^}]+\}|\[[^\]]+\]|\([^)]+\)", "", text)

    # Remove special characters and numbers
    text = re.sub(r"[^a-zA-Z\s'.]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text.lower()
```

# Get stop words

```python
stop_word = []
with open('stop.txt','r') as file:
    content = file.readlines()
    for line in content:
        stop_word.append(line.lower())
stop_word = [x.replace('\n','') for x in stop_word]
```

```python
positive_word = []
with open('positive-words.txt','r') as file:
    content = file.readlines()
    for line in content:
        positive_word.append(line.lower())
positive_word = [x.replace('\n','') for x in positive_word]
```

```python
negative_word = []
with open('negative-words.txt','r') as file:
    content = file.readlines()
    for line in content:
        negative_word.append(line.lower())
negative_word = [x.replace('\n','') for x in negative_word]
```

# Stop Positive Negative

```python
def remove_stop_word(text,stop_word):
    text = clean_text(text)
    ftext = []
    words = text.split(' ')
    for word in words:
        if (word not in stop_word) and (word != '.'):
            word = word.replace('.','')
            ftext.append(word)
    return ' '.join(ftext)
    
```

```python
df['Clean'] = df['Text'].apply(lambda x:remove_stop_word(x,stop_word))
df
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def count_pos(x,positive):
    count = 0
    sen = nlp(x)
    for i in sen:
        if i.lemma_ in positive:
            count += 1
    return count
```

```python
df['Positive Score'] = df['Clean'].apply(lambda x: count_pos(x,positive_word))
df
```

```python
def count_neg(x,negative):
    count = 0
    sen = nlp(x)
    for i in sen:
        if i.lemma_ in negative:
            count += 1
    return count
```

```python
df['Negative Score'] = df['Clean'].apply(lambda x: count_neg(x,negative_word))
df
```

# Polarity

```python
sia = SentimentIntensityAnalyzer()
def get_polarity(text):
    text = clean_text(text)
    return sia.polarity_scores(text)['compound']
```

```python
df['Polarity Score'] = df['Clean'].apply(get_polarity)
df
```

```python
def get_seniment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Polarity Score'].apply(get_seniment)
df
```

# Visualization

```python
df['Sentiment'].value_counts().plot(kind='bar')
plt.show()
```

```python
wc  = WordCloud(width=1000, height=800,background_color='white').generate(" ".join(df[df['Sentiment']=='Neutral']['Clean']))
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word-Cloud For Neutral Sentiment')
plt.show()
```

```python
wc  = WordCloud(width=1000, height=800,background_color='white').generate(" ".join(df[df['Sentiment']=='Positive']['Clean']))
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word-Cloud For Positive Sentiment')
plt.show()
```

```python
wc  = WordCloud(width=1000, height=800,background_color='white').generate(" ".join(df[df['Sentiment']=='Negative']['Clean']))
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word-Cloud For Negative Sentiment')
plt.show()
```
