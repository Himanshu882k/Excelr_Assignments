```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi']= 300
import seaborn as sns
sns.set_theme(style='darkgrid',palette='rainbow')
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report ,confusion_matrix
from sklearn.preprocessing import LabelEncoder , OrdinalEncoder
%matplotlib inline
```

```python
df = pd.read_csv('SalaryData_Train(1).csv')
df
```

### Check for null and duplicate

```python
df.describe()
```

```python
df.isna().sum()
```

```python

df.drop(index=list(df[df.duplicated()].index),inplace=True)
```

```python
df.info()
```

```python
print(f"Non Numeric Columns : {[x for x in df.columns if df[x].dtype=='O']}")
```

```python
lbenc = LabelEncoder()
df['Salary']=lbenc.fit_transform(df['Salary'])
```

```python
oe= OrdinalEncoder()
df[['workclass', 'education', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex', 'native']] = oe.fit_transform(df[['workclass', 'education', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex', 'native']])
```

```python
df
```

# Visualization

```python
sns.countplot(data=df,x='Salary')
```

```python
sns.scatterplot(x=df['maritalstatus'],y=df['education'],hue=df['Salary'])
```

```python
sns.scatterplot(x=df['age'],y=df['hoursperweek'],hue=df['Salary'])
```

## Data Preprocessing

```python
x = df.drop('Salary',axis=1)
y= df['Salary']
```

```python
std = StandardScaler()
scaled =std.fit_transform(x)
scaled
```

### Spliting Data

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.15,random_state=42)
```

# SVC

```python
svm = SVC()
svm.fit(xtrain,ytrain)
```

```python
print(f"Train Score : {svm.score(xtrain,ytrain)}")
print(f"Test Score : {svm.score(xtest,ytest)}")
```

```python
ypred = svm.predict(xtest)
```

```python
print('Train Report')
print(classification_report(ytrain,svm.predict(xtrain)))
```

```python
print(' Test Report')
print(classification_report(ytest,ypred))
```

```python
sns.heatmap(confusion_matrix(ytest,ypred),annot=True,fmt='.2f')
plt.title('Confusion Matrix')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```

# Hyper Parameter Tuning

```python
params = {
    'kernel':['linear','rbf'],
    'gamma':[0.1,0.01],
    'C':[0.01,1]
}
```

```python
grid=GridSearchCV(SVC(),param_grid=params,verbose=1)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
svm.fit(xtrain,ytrain)
ypred = svm.predict(xtest)
```

```python
svm.score(xtrain,ytrain), svm.score(xtest,ytest)
```

# Final Model

```python
fmodel = SVC(C=1,gamma=0.1,kernel='rbf')
```

```python
fmodel.fit(xtrain,ytrain)
```

```python
print(f"Train Score : {fmodel.score(xtrain,ytrain)}")
print(f"Test Score : {fmodel.score(xtest,ytest)}")
```

```python
print(' Train Report')
print(classification_report(ytrain,fmodel.predict(xtrain)))
```

```python
print(' Test Report')
print(classification_report(ytest,ypred))
```

```python
sns.heatmap(confusion_matrix(ytest,ypred),annot=True,fmt='.2f')
plt.title('Confusion Matrix')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```
