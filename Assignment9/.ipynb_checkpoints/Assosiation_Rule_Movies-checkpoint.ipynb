```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi']=250
```

```python
df = pd.read_csv('my_movies.csv')
df
```

```python
mv = df.iloc[:,5:]
mv
```

# Apriori Algorithm

```python
#with 20% support
freq_movie = apriori(mv.astype(bool),min_support=0.2,use_colnames=True)
freq_movie
```

```python
freq_movie.sort_values("support",ascending=False)
```

# Assosiation Rule

```python
#with 70% confidence
rules = association_rules(freq_movie,metric='confidence',min_threshold=0.7)
rules
```

```python
#lift > 1 is good
rules[rules.lift>1]
```

```python
#visualization of rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
```

# Lets try assosiation rule with different threshold value for support and confidence

```python
#with 10% support
freq_movies = apriori(mv.astype(bool),min_support=0.1,use_colnames=True)
freq_movies
```

```python
freq_movies.sort_values("support",ascending=False)
```

```python
#with 80% confidence
rule = association_rules(freq_movies,metric='confidence',min_threshold=0.8)
rule
```

```python
rule[rule.lift>1]
```

```python
#visualization of rule
plt.scatter(rule['support'],rule['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
```
